```python
def detail_feature(dataset,threshold):
    obj = set()
    cor = dataset.corr()
    for i in range(len(dataset.columns)):
        for j in range(i):
            if abs(cor.iloc[i,j]) > threshold:
                obj.add(cor.columns[i])
                
    return obj
```

```python
features = detail_feature(X_train,0.85)
```

```python
features
```

```python
X_train.drop(features,axis=1,inplace=True)
X_test.drop(features,axis=1,inplace=True)
```
